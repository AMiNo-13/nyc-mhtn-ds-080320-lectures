```python
z_critical=stats.norm.ppf(q=0.975) # The Z value (point) on the z distribution 
z_critical
```

```python
stats.norm.interval(0.95) # The two points (z values) on z distribution.
```

#### Standard error:

```python
import math
```

```python
standard_error= sigma/math.sqrt(n)
standard_error
```

```python
CI_lower= x_bar - z_critical * standard_error
CI_upper= x_bar + z_critical * standard_error
CI_lower, CI_upper
```
