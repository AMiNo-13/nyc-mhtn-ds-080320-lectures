## Question 6

Explain what a type 1 error is and how does it relate to the significance level when doing a statistical test. 

Type 1 Error is known as false positive which means that when we accept an alternative hypothesis, but it is actually false. The alpha (significance level)that we pick is the likelihood that we WILL get a type 1 error due to random chance. TYPE 1 Error is when we reject the null hypothesis when we shouldnt have rejected the null hypothesis 
Type 2 Error which is known as false negative is when we reject an alternative hypothesis when it is actually TRUE.
